```python
q=QuantumRegister(3,'q')
anc=QuantumRegister(1,'anc')
nanc=QuantumRegister(1,'nanc')
qc=QuantumCircuit(q,anc,nanc,name='qc')
marked=[1,1,0]
qc.h(q[0])
qc.h(q[1])
qc.h(q[2])
qc.x(anc[0])
oq.Wavefunction(qc,systems=[3,1,1],show_systems=[True,False,False])
iterations=4
for i in range(iterations):
    oq.Grover_Oracle(marked,qc,q,anc,nanc)
    oq.Grover_Diffusion(marked,qc,q,anc,nanc)
    print("Grover algo")
    oq.Wavefunction(qc,systems=[3,1,1],show_systems=[True,False,False])
"""fter 1 Grover Iteration, 
we have a 78% percent chance of measuring our marked state.
 After the second Grover Iteration, this probability jumps to over 94%! 
 But, if we apply a third iteration, our probability of measuring the marked state plummets to a measly 33% (but is still the highest single state). 
aftr fourth iteration, we find our marked state with a 1% probability, 
the complete opposite of what we set out to do!
"""
```

```python
marked=[0,1,1,0]
Q=4
qc,q,anc,nanc,c=oq.Grover(Q,marked)
oq.Wavefunction(qc,systems=[Q,1,Q-2],show_systems=[True,False,True],column=True)#positional grouping parameter specific to the OQ framework
#q-2 denotes remaining qubits
qc.measure(q,c)
oq.Measurement(qc,shots=100)
```
